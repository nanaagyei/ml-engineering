```python
import numpy
import pandas as pd
import seaborn as sn

# import dataset
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(["PassengerId"], axis=1)
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.isnull().sum()
```

```python
df.head()
```

```python
df = df.drop(["Name", "Cabin"], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Age'] = df['Age'].astype(int)
df['Fare'] = df['Fare'].astype(int)
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df["Age"].hist(bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df.boxplot(column="Age", by="Survived")
plt.show()
```

```python
plt.scatter(x="Age", y="Fare", data=df)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
df = df[df["Fare"] < 300]
```

```python
plt.scatter(x="Age", y="Fare", data=df)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
